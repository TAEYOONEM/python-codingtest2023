```python
cnt = 10

def openBox() :
    global cnt
    print("open box")
    cnt -= 1
    if cnt == 0 :
        print("input ring and return")
        return
    openBox()
    print("close box")
```

## 재귀호출 하지않고 반복문으로 계산

```python
sumValue = 0

for n in range(10,0,-1) :
    sumValue += n

print(sumValue)
```

```python
def addNumber(num) :
    if num <= 1 :
        return 1
    else :
        return num + addNumber(num-1)
        
print(addNumber(3))
```

### 팩토리얼 반복문

```python
factValue = 1
for n in range(10,0,-1) :
    factValue *= n
print(factValue)
```

### 팩토리얼 재귀호출

```python
def factorial(num) :
    if num <= 1:
        return 1
    return num * factorial(num-1)

print(factorial(10))
```

### 우주선 발사

```python
import time
def countdown(n) :
    if n == 0 :
        print("발사")
    else :
        print(n)
        time.sleep(1)
        countdown(n-1)

countdown(5)
```

```python
def printStar(n) :
    if n > 0 :
        printStar(n-1)
        print("*" * n)

printStar(5)
```

### 구구단 재귀호출

```python
def gugudan(dan,num) :
    print(f"{dan} x {num} ={dan * num:2d}",end=' ')
    if num < 9 :
        gugudan(dan,num+1)
for dan in range(2,10) :
    print(f"\n{dan}단 시작---")
    gugudan(dan,1)
```

### n제곱

```python
tab = ' '
def pow(x,n) :
    global tab
    tab += ' '
    if n == 0 :
        return 1
    print(tab,f"{x} ^ {n} ") 
    return x * pow(x,n-1) 

print(pow(2,10))   
```

### 피보나치

```python
def fibo(n) :
    if n == 0 :
        return 0
    elif n == 1 :
        return 1

    return fibo(n-1) + fibo(n-2)

print("0 1",end=' ')
for i in range(2, 10) :
    print(fibo(i),end=' ')
```

### 희문여부

```python
def palindrom(pStr) :
    if len(pStr) <= 1 :
        return True
    
    elif pStr[0] != pStr[-1] :
        return False

    return palindrom(pStr[1:len(pStr)-1])
```

```python
strAry = ['reaver','kayak','Borrow or rob', '주유소의 소유주']
for testStr in strAry :
    print(testStr, end="->")
    testStr = testStr.lower().replace(' ','')
    if palindrom(testStr) :
        print('회문입니다.')
    else : 
        print('회문 아니노')
```
